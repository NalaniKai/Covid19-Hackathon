```python
for i in range(len(imperialCollegeData)):
    #print("operating")
    state = imperialCollegeData['StateName'][i]
    start_date = str(imperialCollegeData['DateEnacted'][i])
    end_date = imperialCollegeData['DateExpiry'][i]
    mandate = imperialCollegeData['Mandate'][i]
    statewide = imperialCollegeData['StateWide'][i]
    policy = imperialCollegeData['StatePolicy'][i]
    policyMandate = policy + "Mandate"
    policyStateWide = policy + "StateWide"
    if(policy not in policies):
        continue
    if(pd.isna(end_date)):
        end_date = "20200612"
    else:
        end_date = str(end_date)
    start_date = datetime(year=int(start_date[0:4]), month=int(start_date[4:6]), day=int(start_date[6:8]))
    end_date = datetime(year=int(end_date[0:4]), month=int(end_date[4:6]), day=int(end_date[6:8]))
    if(end_date > datetime(year=int(2020),month=int(6),day=int(12))):
        end_date = datetime(year=int(2020),month=int(6),day=int(12))
    #print(start_date)
    #s = start_date.strftime("%Y-%m-%d")
    while(start_date < end_date):
        index = result.index[(result['date'] == start_date.strftime("%Y-%m-%d")) & (result['state'] == state)].tolist()
        if(len(index)==0):
            print("Not in NY Times, Imperial College Index at: " + str(i))
            print("Date: " + start_date.strftime("%Y-%m-%d"))
            print("State: " + state)
            start_date += timedelta(days=1)
            continue
        else:
            index = index[0]
        result.at[index, policy] = 1
        result.at[index, policyMandate] = mandate
        result.at[index, policyStateWide] = statewide
        #print("num of col: " + str(len(result.loc[index])))
        #result.loc[index,policy] = 1
        start_date += timedelta(days=1)
        
        
        
result
```
