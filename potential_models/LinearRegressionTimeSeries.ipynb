Linear Regression Time Series

```python
import pandas

data_path = "../data/"
filename = "combined_data.csv"

df = pandas.read_csv(data_path + filename)
df = df.drop(['Unnamed: 0'], axis=1)
df = df.drop(['fips'], axis=1)
df.head()
```

```python
target = "cases" #"Deaths", "Cases"
not_features = ["date", "state", "deaths"]
not_features.append(target)

target_values = df[target]
feature_values = df

for x in not_features:
    feature_values = feature_values.drop(x, axis=1)

feature_values.head()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression



lm = LinearRegression()

result = lm.fit(feature_values,target_values)
print(result.intercept_)
print(result.coef_)

```
